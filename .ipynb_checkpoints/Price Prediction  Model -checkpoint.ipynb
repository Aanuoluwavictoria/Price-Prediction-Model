```python
# Let's transform 'Price' with a log transformation
log_price=np.log(data_cleaned['Price'])
# The we add it to our data frame
data_cleaned['log_price']=log_price
data_cleaned
```

```python
# let check the three scatter plot once again
F,(ax1,ax2,ax3)=plt.subplots(1,3,sharey=True,figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log price and Mileage')
plt.show
```

```python
# The relationship show a clear linear relationship
# This is some good linear regression material
#Alternatively we could have transformed each of the independet variable

# Since we will be using the log price variable wecan drop the old 'price' one
data_cleaned=data_cleaned.drop(['Price'],axis=1)
data_cleaned
```

# Create dummy Variables

```python
# To include the categorical data in the in the regression,lets create dummies
# There is a very conveient method called: get_dummies which does that seemlessly
# It is extremely important that drop one of the dummies,alternatively we will introduce multically
data_with_dummies=pd.get_dummies(data_no_multheollinarity,drop_first=True) 
#pd.get_dummies(df[.drop_first])
```
